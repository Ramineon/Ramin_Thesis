```python
import numpy as np
import matplotlib.pyplot as plt

# Replace with the path to your .npy file
npy_file_path = r"G:\Thesis_Numpy_data_set\2_Class_320\1\ALL0001.CSV_iter5.npy"

# Load the NumPy array
sample_array = np.load(npy_file_path)

# Determine the figure size based on array shape
rows, cols = sample_array.shape
fig_size = (cols / 100, rows / 100)  # Scale to keep dimensions consistent

# Create the plot
plt.figure(figsize=fig_size, dpi=100)
plt.imshow(sample_array, cmap='gray')
plt.axis('off')  # Hide axis
plt.title(f"Pure RP pic ({rows}x{cols})")

# Save the image as a file
output_image_path = r"output_image.png"  # Change to desired file name and location
plt.savefig(output_image_path, dpi=rows)  # DPI matches rows to ensure pixel-perfect size
plt.show()
```
